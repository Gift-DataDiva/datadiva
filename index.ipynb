```python
import pandas as pd
```

```python
df=pd.read_csv(r"food industry dataset 1.csv")
```

```python
df
```

```python
df.info()
```

```python
df["Menu_Item"]=df["Menu_Item"].fillna("pizza")
```

```python
df
```

```python
df.info()
```

```python
df["Delivery_Date"]=df["Delivery_Date"].fillna("2/22/2023")
```

```python
df
```

```python
df["Unit_Price"]=df["Unit_Price"].fillna("11.4")
```

```python
df
```

```python
df.info()
```

```python
df["Inspection_Status"]=df["Inspection_Status"].fillna("Pending")
```

```python
df
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
```

```python
val =df["Total_Cost"].max()/6
```

```python
bins=[0,val,val*2,val*3,val*4,val*5,val*6]
labels=["<100k","100-500k","500-1m","1m-5m","5m-10m","10m"]
```

```python
val*4
```

```python
df["cost_bucket"]=pd.cut(df["Total_Cost"],bins=bins, labels=labels, include_lowest=True)
```

```python
df
```

```python
Total_Cost_sum = df.groupby("cost_bucket", observed=True)["Total_Cost"].sum()
```

```python
Total_Cost_sum
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(Total_Cost_sum.index.astype(str), Total_Cost_sum.values,color="skyblue")
plt.title("Total Total Cost by cost Bucket")
plt.xlabel("Cost Bucket")
plt.ylabel("Total Total Cost")
plt.xticks(rotation=45)
plt.grid(True, axis="y")
plt.tight_layout()
plt.show()
```

```python
val =df["Unit_Price"].astype(float).max()/8
```

```python
val
```

```python
df["Unit_Price"] = df["Unit_Price"].astype(float)
```

```python
df.info()
```

```python
bins=[0,val,val*2,val*3,val*4,val*5,val*6,val*7,val*8]
labels=["<10k","10-50k","50-100k","100-500k","500k-1m","1-5m","5-10m","50m"]
```

```python
df["cost_name"]=pd.cut(df["Unit_Price"],bins=bins, labels=labels, include_lowest=True)
```

```python
df
```

```python
Unit_Price_sum = df.groupby("cost_name",observed=True)["Unit_Price"].sum()
```

```python
Unit_Price_sum
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(Unit_Price_sum.index.astype(str), Unit_Price_sum.values,color="skyblue")
plt.title("Total gift shop by cost name")
plt.xlabel("Cost Name")
plt.ylabel("Total gift shop")
plt.xticks(rotation=45)
plt.grid(True, axis="y")
plt.tight_layout()
plt.show()
```
